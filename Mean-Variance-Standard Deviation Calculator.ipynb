Mean-Variance-Standard Deviation Calculator

```python
import numpy as np

def calculate(list):
    if (len(list) != 9):
      raise ValueError("List must contain nine numbers.")
    # split into 3x3 matrix [[1, 2, 3], [1, 2, 3], [1, 2, 3]]
    flat_arr = np.array(list)
    arr = np.array(list).reshape((3, 3))
    # get col, row, and flat average
    col_avg = np.mean(arr, axis=0)
    row_avg = np.mean(arr, axis=1)
    flat_avg = np.mean(flat_arr)
    # get col, row, and flat variance
    col_var = np.var(arr, axis=0)
    row_var = np.var(arr, axis=1)
    flat_var = np.var(flat_arr)
    # get col, row, and flat stdev
    col_std = np.std(arr, axis=0)
    row_std = np.std(arr, axis=1)
    flat_std = np.std(flat_arr)
    # get col, row, and flat maximum
    col_max = arr.max(axis=0)
    row_max = arr.max(axis=1)
    flat_max = arr.max()
    # get col, row, and flat minimum
    col_min = arr.min(axis=0)
    row_min = arr.min(axis=1)
    flat_min = arr.min()
    # get col, row, and flat sum
    col_sum = []
    # for every col
    for i in range(3):
        currColSum = 0;
        # for every row
        for j in range(3):
            currElem = arr[j][i]
            currColSum += currElem
        col_sum.append(currColSum)
    row_sum = [np.sum(x) for x in arr]
    flat_sum = np.sum(list)
    # render as dictionary
    mean_arr = [col_avg.tolist(), row_avg.tolist(), flat_avg.tolist()]
    var_arr = [col_var.tolist(), row_var.tolist(), flat_var.tolist()]
    std_arr = [col_std.tolist(), row_std.tolist(), flat_std.tolist()]
    max_arr = [col_max.tolist(), row_max.tolist(), flat_max.tolist()]
    min_arr = [col_min.tolist(), row_min.tolist(), flat_min.tolist()]
    sum_arr = [col_sum, row_sum, flat_sum]
    calculations = {'mean': mean_arr, 'variance': var_arr, 'standard deviation': std_arr,
                    'max': max_arr, 'min': min_arr, 'sum': sum_arr}
    return calculations
```

```python
calculate([0,1,2,3,4,5,6,7,8])
```

```python
calculate([2,6,2,8,4,0,1,5,7])
```
