```python
from google.colab import drive
drive.mount('/content/gdrive')
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
```

```python
df = pd.read_csv('gdrive/My Drive/Colab Notebooks/csvs/epa-sea-level.csv')
```

```python
x_vals = np.array(df['Year'].to_numpy().tolist()[113:])
new_x_vals = np.array(df['Year'].to_numpy().tolist()[121:])
y_vals = np.array(df['NOAA Adjusted Sea Level'].to_numpy().tolist()[113:])
new_y_vals = np.array(df['NOAA Adjusted Sea Level'].to_numpy().tolist()[121:])

old_fit = linregress(x_vals, y_vals)
new_fit = linregress(new_x_vals, new_y_vals)
y_old = old_fit.intercept + old_fit.slope * x_vals
y_new = new_fit.intercept + new_fit.slope * new_x_vals

plt.xlim(left=1992, right=2050)
plt.ylim(bottom=5, top=14)
plt.scatter(x_vals, y_vals)

x = np.linspace(1992, 2050)
plt.plot(x, old_fit.intercept + old_fit.slope * x, color='red', label='regression line')
plt.plot(x, new_fit.intercept + new_fit.slope * x, color='green', label='regression line')

plt.title('Rise in Sea Level')
plt.xlabel('Year')
plt.ylabel('NOAA Adjusted Sea Level')
plt.show()
```
