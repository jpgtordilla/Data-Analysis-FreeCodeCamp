```python
import pandas as pd

def calculate_demographic_data(print_data=True):
    # Read data from file
    df = pd.read_csv('gdrive/My Drive/Colab Notebooks/csvs/adult.data.csv')

    def getRaceList():
      white_list = [x for x in df['race'].tolist() if x == 'White']
      black_list = [x for x in df['race'].tolist() if x == 'Black']
      asian_list = [x for x in df['race'].tolist() if x == 'Asian-Pac-Islander']
      nat_list = [x for x in df['race'].tolist() if x == 'Amer-Indian-Eskimo']
      other_list = [x for x in df['race'].tolist() if x == 'Other']
      races = [white_list, black_list, asian_list, nat_list, other_list]
      race_list = []
      for race in races:
        race_list.append(len(race))
      return race_list

    # How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
    race_count = getRaceList()

    def getAvgMenAge():
      # delete rows with not men
      men_df = df[df.sex == "Male"]
      # get average age
      avg = np.mean(men_df["age"].to_numpy())
      return round(avg, 1)

    # What is the average age of men?
    average_age_men = getAvgMenAge()

    def getPercentageBachelors():
      bachelordf = df[df.education == "Bachelors"]
      ratio = len(bachelordf) / len(df)
      return round(ratio * 100, 1)

    # What is the percentage of people who have a Bachelor's degree?
    percentage_bachelors = getPercentageBachelors()

    # What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
    # What percentage of people without advanced education make more than 50K?

    # with and without `Bachelors`, `Masters`, or `Doctorate`

    def getPercentageRichHigher():
      bachelor_df = df[df.education == "Bachelors"]
      master_df = df[df.education == "Masters"]
      doctor_df = df[df.education == "Doctorate"]
      high_sum = len(bachelor_df) + len(master_df) + len(doctor_df)
      num_higher_50 = 0
      bach_list = bachelor_df['salary'].to_numpy().tolist()
      for salary in bach_list:
        if salary == '>50K':
          num_higher_50 += 1
      master_list = master_df['salary'].to_numpy().tolist()
      for salary in master_list:
        if salary == '>50K':
          num_higher_50 += 1
      doctor_list = doctor_df['salary'].to_numpy().tolist()
      for salary in doctor_list:
        if salary == '>50K':
          num_higher_50 += 1

      ratio = num_higher_50 / high_sum
      return round(ratio * 100, 1)

    def getPercentageRich():
      no_bachelors = df.drop(df[df['education'] == 'Bachelors'].index)
      no_m_b = no_bachelors.drop(no_bachelors[no_bachelors['education'] == 'Masters'].index)
      no_higher = no_m_b.drop(no_m_b[no_m_b['education'] == 'Doctorate'].index)
      low_sum = len(no_higher)
      num_higher_50 = 0
      low_list = no_higher['salary'].to_numpy().tolist()
      for salary in low_list:
        if salary == '>50K':
          num_higher_50 += 1

      ratio = num_higher_50 / low_sum
      return round(ratio * 100, 1)

    def getPercentageHigher():
      bachelor_df = df[df.education == "Bachelors"]
      master_df = df[df.education == "Masters"]
      doctor_df = df[df.education == "Doctorate"]
      high_sum = len(bachelor_df) + len(master_df) + len(doctor_df)
      ratio = high_sum / len(df)
      return round(ratio * 100, 1)

    def getPercentageLower():
      percentHigh = getPercentageHigher()
      return round(100 - percentHigh, 1)

    higher_education = getPercentageRichHigher()
    lower_education = getPercentageLower()

    # percentage with salary >50K
    higher_education_rich = getPercentageRichHigher()
    lower_education_rich = getPercentageRich()

    # What is the minimum number of hours a person works per week (hours-per-week feature)?
    min_work_hours = min(df['hours-per-week'].to_numpy())

    def richMinWorkers():
      min_hour_df = df[df['hours-per-week'] == 1]
      salary_list = min_hour_df['salary'].to_numpy().tolist()
      num_people_rich = 0
      for salary in salary_list:
        if salary == '>50K':
          num_people_rich += 1
      ratio = num_people_rich / len(salary_list)
      return round(ratio * 100, 1)

    # What percentage of the people who work the minimum number of hours per week have a salary of >50K?
    num_min_workers = len(df[df['hours-per-week'] == 1])

    rich_percentage = richMinWorkers()

    def getHighestCountry():
      countries = set(df['native-country'].to_numpy().tolist()[:])
      countries.remove('?')
      # generate country dfs and store in a list
      country_dfs = []
      for country in countries:
        current_df = df[df['native-country'] == country]
        country_dfs.append(current_df)

      # generate country salary dictionary with percentages rich

      country_wealth_dict = {}
      for country_df in country_dfs:
        # get percentage above 50K, add percentage to dictionary
        current_count = 0
        current_salary_list = country_df['salary'].to_numpy().tolist()
        for salary in current_salary_list:
          if salary == '>50K':
            current_count += 1
        ratio_rich = current_count / len(country_df)
        percentage_rich = round(100 * ratio_rich, 1)
        country_wealth_dict.update({country_df['native-country'].to_numpy().tolist()[0]: percentage_rich})
      # get country with highest percentage
      return max(country_wealth_dict, key=country_wealth_dict.get)

    def getHighestCountryPercent():
      countries = set(df['native-country'].to_numpy().tolist()[:])
      countries.remove('?')
      # generate country dfs and store in a list
      country_dfs = []
      for country in countries:
        current_df = df[df['native-country'] == country]
        country_dfs.append(current_df)

      # generate country salary dictionary with percentages rich

      country_wealth_dict = {}
      for country_df in country_dfs:
        # get percentage above 50K, add percentage to dictionary
        current_count = 0
        current_salary_list = country_df['salary'].to_numpy().tolist()
        for salary in current_salary_list:
          if salary == '>50K':
            current_count += 1
        ratio_rich = current_count / len(country_df)
        percentage_rich = round(100 * ratio_rich, 1)
        country_wealth_dict.update({country_df['native-country'].to_numpy().tolist()[0]: percentage_rich})
      # get country with highest percentage
      return max(country_wealth_dict.values())

    # What country has the highest percentage of people that earn >50K?
    highest_earning_country = getHighestCountry()
    highest_earning_country_percentage = getHighestCountryPercent()

    # Identify the most popular occupation for those who earn >50K in India.

    def getPopularIndiaJob():
      india_df = df[df['native-country'] == 'India']
      rich_india_df = india_df[india_df['salary'] == '>50K']
      occupation_list = rich_india_df['occupation'].to_numpy().tolist()
      occupation_set = set(occupation_list)
      occupation_dict = {}
      for occupation in occupation_set:
        # get the number of people with that occupation, add occupation to dictionary
        num_with_occupation = 0
        for person in occupation_list:
          if person == occupation:
            num_with_occupation += 1
        occupation_dict.update({occupation: num_with_occupation})
      return max(occupation_dict, key=occupation_dict.get)

    top_IN_occupation = getPopularIndiaJob()

    # DO NOT MODIFY BELOW THIS LINE

    if print_data:
        print("Number of each race:\n", race_count)
        print("Average age of men:", average_age_men)
        print(f"Percentage with Bachelors degrees: {percentage_bachelors}%")
        print(f"Percentage with higher education that earn >50K: {higher_education_rich}%")
        print(f"Percentage without higher education that earn >50K: {lower_education_rich}%")
        print(f"Min work time: {min_work_hours} hours/week")
        print(f"Percentage of rich among those who work fewest hours: {rich_percentage}%")
        print("Country with highest percentage of rich:", highest_earning_country)
        print(f"Highest percentage of rich people in country: {highest_earning_country_percentage}%")
        print("Top occupations in India:", top_IN_occupation)

    return {
        'race_count': race_count,
        'average_age_men': average_age_men,
        'percentage_bachelors': percentage_bachelors,
        'higher_education_rich': higher_education_rich,
        'lower_education_rich': lower_education_rich,
        'min_work_hours': min_work_hours,
        'rich_percentage': rich_percentage,
        'highest_earning_country': highest_earning_country,
        'highest_earning_country_percentage':
        highest_earning_country_percentage,
        'top_IN_occupation': top_IN_occupation
    }
```

```python
calculate_demographic_data()
```
